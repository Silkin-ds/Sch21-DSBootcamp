```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../../datasets/ab-test.csv')
df
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 8))
df[df['time'] == 'after'].boxplot(column='diff',
                                  by='group',
                                  ax=ax[0],
                                  color=dict(whiskers='red',
                                             medians='red', caps='red'),
                                  boxprops=dict(color = 'green', linewidth=3),
                                  medianprops=dict(linewidth=2),
                                  showmeans=True)
df[df['time'] == 'before'].boxplot(column='diff',
                                   by='group',
                                   ax=ax[1],
                                   color=dict(whiskers='red',
                                              medians='red', caps='red'),
                                   boxprops=dict(color = 'green', linewidth=3),
                                   medianprops=dict(linewidth=2),
                                   showmeans=True)
ax[0].set_title('after')
ax[1].set_title('before')
fig.suptitle(
    'Lag before labs deadlines. After and before the Newsfeed', fontsize=15)
plt.show()
```

IQR контрольной группы до появления ленты новостей примерно 150, а после появления ленты новостей примерно 70
