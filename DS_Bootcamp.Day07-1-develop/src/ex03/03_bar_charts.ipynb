```python
import pandas as pd
import sqlite3
```

```python
connect = sqlite3.connect('../../datasets/checking-logs.sqlite')
```

```python
query = """
SELECT uid,timestamp FROM checker WHERE uid LIKE 'user_%'
"""
```

```python
commits = pd.read_sql(query,connect,parse_dates=['timestamp'])
commits
```

```python
commits['weekday'] = commits['timestamp'].dt.dayofweek
commits['day'] = commits.timestamp.dt.day
commits['hour'] = commits.timestamp.dt.hour
commits['type'] = commits['weekday'].apply(
    lambda x: 'working_day' if x < 5 else 'weekend')
commits = commits.drop(columns = ['weekday'])
commits
```

```python
commits_groupped = commits.groupby(['type', 'day', 'hour']).count()
commits_groupped = commits_groupped.groupby(
    ['type', 'hour']).mean()['uid'].reset_index()
commits_groupped
```

```python
pivot_table = pd.pivot(commits_groupped,
                       values='uid',
                       index='hour',
                       columns='type')
pivot_table
```

```python
full_hours = pd.Index(range(24),name='hour')
pivot_table = pivot_table.reindex(full_hours)
print(pivot_table)
```

```python
pivot_table.plot(kind='bar',
                 figsize=(15,8),
                 title='Commits per hour',
                 rot=90,
                 fontsize=8,
                 color=['blue', 'red'],
                 subplots=True)
```
