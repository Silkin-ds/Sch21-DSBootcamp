считываем json из ex02 и определяем формат для одного из столбцов

```python
df = pd.read_json('../../datasets/auto.json')
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

создаем выборку из 200 новых наблюдений с random_state = 21 и объединяем с исходным df и сохраняем в новый concat_rows

```python
np.random.seed(21)
valid = df[['CarNumber', 'Make', 'Model']].drop_duplicates()
sample = valid.sample(n=200, replace = True, random_state = 21)
sample_fines = df['Fines'].sample(n=200, replace = True, random_state = 21).values
sample_refund = df['Fines'].sample(n=200, replace = True, random_state = 21).values
sample = pd.DataFrame({'CarNumber': sample['CarNumber'].values,
                            'Make': sample['Make'].values,
                            'Model': sample['Model'].values,
                            'Fines': sample_fines,
                            'Refund': sample_refund})

concat_rows = pd.concat([df,sample]).reset_index(drop=True)
```

```python
concat_rows.head(1000)
```

создаем столбец year и соединяем с df

```python
np.random.seed(21)
concat_rows['Year'] = np.random.randint(1980, 2020, size=len(concat_rows))
fines = concat_rows
```

создаем совмещенный df

```python
surnames = pd.read_json('../../datasets/surname.json')
surnames.columns = surnames.iloc[0]
surnames = surnames.iloc[1:]
surnames = surnames.reset_index(drop=True)
np.random.seed(21)
unique_nums = concat_rows['CarNumber'].unique()
r_surnames = np.random.choice(surnames['NAME'], size = len(unique_nums), replace = True)
owners = pd.DataFrame({'CarNumber': unique_nums,
                       'SURNAME': r_surnames})
owners
```

```python
new_fines_data = [
    {'CarNumber': 'XYZ123RUS', 'Make': 'Tesla', 'Model': 'Model 3', 'Refund': 2.0, 'Fines': 0.0, 'Year': 2002},
    {'CarNumber': 'ABC456RUS', 'Make': 'BMW', 'Model': 'X5', 'Refund': 1.0, 'Fines': 7500.0, 'Year': 2009},
    {'CarNumber': 'DEF789RUS', 'Make': 'Audi', 'Model': 'A4', 'Refund': 1.0, 'Fines': 3200.0, 'Year': 2002},
    {'CarNumber': 'GHI012RUS', 'Make': 'Mercedes', 'Model': 'C-Class', 'Refund': 2.0, 'Fines': 1500.0, 'Year': 2007},
    {'CarNumber': 'JKL345RUS', 'Make': 'Volvo', 'Model': 'XC90', 'Refund': 1.0, 'Fines': 4200.0, 'Year': 2018}
]
fines = pd.concat([fines, pd.DataFrame(new_fines_data)], ignore_index=True)
owners = owners.iloc[:-20]
new_owners_data = [
    {'CarNumber': 'MNO678RUS', 'SURNAME': 'NOVAK'},
    {'CarNumber': 'PQR901RUS', 'SURNAME': 'PETROV'},
    {'CarNumber': 'STU234RUS', 'SURNAME': 'SIDOROV'}
]
owners = pd.concat([owners, pd.DataFrame(new_owners_data)], ignore_index=True)
print(fines.tail())
print(owners.tail())
```

```python
inner_merged = pd.merge(fines, owners, on='CarNumber', how='inner')
outer_merged = pd.merge(fines, owners, on='CarNumber', how='outer')
left_merged = pd.merge(fines, owners, on='CarNumber', how='left')
right_merged = pd.merge(fines, owners, on='CarNumber', how='right')
inner_merged
```

```python
pivot_table = pd.pivot_table(
    data=fines,
    values='Fines',
    index=['Make', 'Model'],
    columns='Year',
    aggfunc='sum',
    fill_value=np.nan
)
formatted_pivot = pivot_table.map(lambda x: f"{x:,.2f}" if pd.notna(x) else "nan")
fines.to_csv('../../datasets/fines.csv', index=False)
owners.to_csv('../../datasets/owners.csv', index=False)
formatted_pivot
```
